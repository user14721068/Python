```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# INTRODUCING THE PYPLOT INTERFACE

print("----Importar la librer√≠a .pyplot")
import matplotlib.pyplot as plt

# Creaci√≥n del objeto fig y ax
# fig   Contenedor donde se guardara todo lo de la pagina.
# ax    Parte de la pagina que tiene los datos. Como un soporte donde se trazara para almacenar los datos.
print("----Metodo .show() para mostrar el contenido de la pagina")
fig, ax = plt.subplots()
plt.show()


# ADDING DATA TO AXIS

print("----Metodo .plot() para graficas Mes vs Temperatura Promedio de Seattle")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],seattle_weather["MLY-TAVG-NORMAL"])
plt.show()

print("----Metodo .plot() para graficas Mes vs Temperatura Promedio de Austin")
fig, ax = plt.subplots()
ax.plot(austin_weather["MONTH"],austin_weather["MLY-TAVG-NORMAL"])
plt.show()


# ADDING MORE DATA IN SAME PLOT

print("----Metodo .plot() para graficas Mes vs Temperatura Promedio de Seattle y Austin")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"],seattle_weather["MLY-TAVG-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"])
plt.show()
```

#     Using the matplotlib.pyplot interface 

There are many ways to use Matplotlib. 
In this course, we will focus on the pyplot interface, which provides the most flexibility in creating and customizing data visualizations.

Initially, we will use the pyplot interface to create two kinds of objects: Figure objects and Axes objects.

This course introduces a lot of new concepts, so if you ever need a quick refresher, download the Matplotlib Cheat Sheet and keep it handy!

#     Customizing your plots (Video)

```python
# CHOOSING MARKERS 

print("----Metodo .plot() y argumento 'marker' para modificar estilo de linea")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker = "o")
plt.show()

# SETTING THE LINESTYLE

print("----Metodo .plot() y argumento linestyle para modificar estilo de linea")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker = "o", linestyle="--")
plt.show()


# ELIMINATING LINES
print("----Metodo .plot() y argumento linestyle para eliminar estilo de linea")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o", linestyle="None")
plt.show()

# CHOOSING COLOR
print("----Metodo .plot() y argumento linestyle para cambiar estilo de grafica")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o", linestyle="--", color="r")
plt.show()

# AXES LABELS
print("----Metodo .set_xlabel() y .set_ylabel() para agregar titulo al eje x")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o", linestyle="--", color="r")
ax.set_xlabel("Tiempo (En Meses)")
ax.set_ylabel("Temperatura Promedio (Grados Farenheit)")
plt.show()

# ADDING A TITLE
print("----Metodo .set_tittle para agregar titulo a la graficas")
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"], marker="o", linestyle="--")
ax.set_xlabel("Tiempo (En Meses)")
ax.set_ylabel("Temperatura Promedio (Grados Farenheit)")
ax.set_title("Temperatura en Seattle")
plt.show()
```

# Plotting time-series with different variables (Video)

```python
# PLOTTING TWO TIME SERIES TOGHETER

print("----Grafica de 2 series de tiempo. Tiene un problema con la escala del eje y")
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.plot(climate_change.index, climate_change["relative_temp"])
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm) / Relative Temperature ")
plt.show()

# USING TWIN AXES
print("----Metodo .twinx() para graficar 2 series de tiempo sobre el mismo eje x, pero con distintas escalas de eje y")
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"])
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm)")

ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"])
ax2.set_xlabel("Time")
ax2.set_ylabel("Relative temperature (Celsius)")
plt.show()


# SEPARATING VARIABLES BY COLOR
print("----Metodo .twinx() para graficar 2 series de tiempo sobre el mismo eje x, pero con distintas escalas de eje y, y con colores distintos")
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color= "blue")
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm)", color = "blue")

ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"], color ="red")
ax2.set_ylabel("Relative temperature (Celsius)", color = "red")
plt.show()


# COLORING THE TICKS
print("----Metodo .tick_params() para modificar el color de los ejes")
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change["co2"], color="blue")
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm)", color="blue")
ax.tick_params('y',colors="blue")


ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"],color="red")
ax2.set_ylabel("Relative temperature (Celsius)", color="red")
ax2.tick_params('y', colors="red")
plt.show()


# A FUNCTION THAT PLOTS TIME SERIES
print("----Una funci√≥n para graficar series de tiempo")

def plot_timeseries(axes, x,y, color, xlabel, ylabel):
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel,color=color)
    axes.tick_params("y",colors=color)
    
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change["co2"],"blue","Time","CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red","Time","Relative temperatures (Celsius)")
plt.show()
    
```

# Defining a function that plots time-series data

```python
print("----Define a function called plot_timeseries that takes as input an Axes object (axes), data (x,y), a string with the name of a color and strings for x- and y-axis labels")
print("----Plot y as a function of in the color provided as the input color")
print("----Set the x- and y-axis labels using the provided input xlabel and ylabel, setting the y-axis label color using color")
print("----Set the y-axis tick parameters using the tick_params method of the Axes object, setting the colors key-word to color")

def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params("y", colors=color)
```

# Using a plotting function

```python
print("----In the provided ax object, use the function plot_timeseries to plot the 'co2' column in blue, with the x-axis label 'Time (years)' and y-axis label 'CO2 levels'")
fig, ax = plt.subplots()
plot_timeseries(ax,  climate_change.index, climate_change["co2"], "blue", "Time", "CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red","Time", "Relative temperature (Celsius)")
plt.show()
```

# Annotating time-series data (Videos)

```python
# ANNOTATIONS 
print("----Metodo .annotate() para agregar una nota en coordenada (2015-10-06, 1) del eje y en rojo")
fig, ax = plt.subplots()
plot_timeseries(ax,  climate_change.index, climate_change["co2"], "blue", "Time", "CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red","Time", "Relative temperature (Celsius)")

ax2.annotate("Mi Texto", xy=(pd.Timestamp("2015-10-06"), 1) )
plt.show()


# POSITIONING THE TEXT
print("----Metodo .annotate() para agregar una nota en cordenada (2015-10-06, 1) y colocar el texto en coordenada (2008-10-06, -0.2)")
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change["co2"], "blue", "Time", "CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red", "Time", "Relative temperature (Celsius)")

ax2.annotate("Mi texto", xy = (pd.Timestamp("2015-10-06"),1) , xytext=(pd.Timestamp("2008-10-06"),-0.2 )  )
plt.show() 


# CONNECTING TEXT TO ANNOTATION WITH ARROW
print("----Metodo .annotate() para conectar con una flecha el texto con una anotaci√≥n")
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change["co2"], "blue", "Time", "CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2,climate_change.index, climate_change["relative_temp"], "red", "Time", "Relative temperature (Celsius)")

ax2.annotate("Mi texto", 
             xy=(pd.Timestamp("2018-10-06"),1), 
             xytext=(pd.Timestamp("2008-10-06"),-0.2),
            arrowprops={} )
plt.show()

# CUSTOMIZING ARROW PROPERTIES
print("----Metodo .annotate() para conectar con una flecha el texto con una anotaci√≥n, y modificar las propiedades de la flecha")
fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change["co2"], "blue", "Time", "CO2 (ppm)")
ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change["relative_temp"], "red", "Time", "Relative temperature (Celsius)")

ax2.annotate("Mi Texto", 
             xy=(pd.Timestamp("2015-10-06"),1),
             xytext=(pd.Timestamp("2008-10-06"),-0.2),
             arrowprops={"arrowstyle":"->", "color":"purple"}
            )

plt.show()
```

#     Quantitative comparisons: bar-charts (Video)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


seattle_weather = pd.read_excel("seattle_weather.xlsx")
austin_weather = pd.read_excel("austin_weather.xlsx")
climate_change = pd.read_excel("climate_change.xlsx")
medals = pd.read_excel("medals.xlsx")
mens_gymnastics = pd.read_excel("mens_gymnastics.xlsx")
mens_rowing = pd.read_excel("mens_rowing.xlsx")
summer_2016_medals = pd.read_excel("summer_2016_medals.xlsx")


#Establecer el indice de cada Data Frame
seattle_weather = seattle_weather.set_index("DATE")
austin_weather = austin_weather.set_index("DATE")
climate_change = climate_change.set_index("date")
medals = medals.set_index("index")
```

```python
# VISUALIZING THE DATA
print("----Metodo .bar() para grafica de barras")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
plt.show()

# ROTATE THE TICK LABELS
print("----Metodo .set_cticklabels() para rotar los valores de los ejes")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel("Numero de medallas")
plt.show()

# STACK BAR CHAR
print("----Metodo .bar() para crear grafica de barras apiladas con 2 indicadores")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.bar(medals.index, medals["Silver"], bottom=medals["Gold"])
ax.set_xticklabels(medals.index, rotation = 90)
ax.set_ylabel("Numero de medallas")
plt.show()

print("----MEtodo .bar() para crear grafica de barras apiladas con 3 indicadores")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.bar(medals.index, medals["Silver"], bottom= medals["Gold"])
ax.bar(medals.index, medals["Bronze"], bottom= medals["Gold"] + medals["Silver"])
ax.set_xticklabels(medals.index, rotation= 90)
ax.set_ylabel("Numero de Medallas")
plt.show()

# AGREGAR ETIQUETAS A LOS COLORES
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.bar(medals.index, medals["Silver"], bottom= medals["Gold"])
ax.bar(medals.index, medals["Bronze"], bottom= medals["Gold"] + medals["Silver"])
ax.set_xticklabels(medals.index, rotation= 90)
ax.set_ylabel("Numero de medallas")
ax.legend(["Gold", "Silver", "Bronze"])
plt.show()
```

# Bar chart

```python
print("----Call the ax.bar method to plot the 'Gold' column as a function of the country")
print("----Use the ax.set_xticklabels to set the x-axis tick labels to be the country names")
print("----In the call to ax.set_xticklabels rotate the x-axis tick labels by 90 degrees by using the rotation key-word argument")
print("----Set the y-axis label to 'Number of medals'")
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"])
ax.set_xticklabels(medals.index, rotation= 90)
ax.set_ylabel("Number of medals")
plt.show()
```

# Stacked bar chart

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals["Gold"], label = "Gold")
ax.bar(medals.index, medals["Silver"], bottom= medals["Gold"], label = "Silver")
ax.bar(medals.index, medals["Bronze"], bottom= medals["Gold"] + medals["Silver"], label = "Bronze")
ax.set_ylabel("Number of medals")
ax.set_xticklabels(medals.index, rotation= 90)
ax.legend()
plt.show() 
```

#     Quantitative comparisons: histograms (Video)

```python
# BAR CHAR AGAIN
print("----Metodo .bar() para mostrar una grafica de barras")
fig, ax = plt.subplots()
ax.bar("Rowing", mens_rowing["Height"].mean() )
ax.bar("Gymnastics", mens_gymnastics["Height"].mean() )
plt.show()

# INTRODUCING HISTOGRAMS
print("----Metodo .hist() para graficar histogramas")
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"])
ax.hist(mens_gymnastics["Height"])
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
plt.show()

# LABELS IN HISTOGRAMS
print("----Argumento label para mostrar etiquetas")
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing")
ax.hist(mens_gymnastics["Height"], label="Gymnastic")
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()

# SETTING THE NUMBER OF BINS IN HISTOGRAM
print("----Argumento bins para elegir el n√∫mero de bins de un histograma")
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", bins=5)
ax.hist(mens_gymnastics["Height"], label="Gymnastic", bins=5)
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()

# SETTING BIN INTERVALS
print("----Argumento bins para elegir los intervalos de los bins de un histograma")
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", bins=[150, 160, 170, 180, 190, 200, 210])
ax.hist(mens_gymnastics["Height"], label="Gymnastic", bins=[150, 160, 170, 180, 190, 200, 210])
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
ax.legend()
plt.show()

# TRANSPERENCY OF HISTOGRAM
print("----Argumento histtype para elegir el tipo de histograma")
fig, ax = plt.subplots()
ax.hist(mens_rowing["Height"], label="Rowing", bins=[150,160,170,180,190,200,210], histtype="step")
ax.hist(mens_gymnastics["Height"], label="Gymnastic", bins=[150,160,170,180,190,200,210], histtype="step")
ax.legend()
ax.set_xlabel("Height (cm)")
ax.set_ylabel("# of observations")
plt.show()
```

#     Statistical plotting (Video)

```python
# ADDING ERROR BARS TO CHARS
print("----Metodo .bar para crear una grafica de barras + marcas de error")
fig, ax = plt.subplots()
ax.bar("Rowing",     mens_rowing["Height"].mean(),     yerr= mens_rowing["Height"].std() )
ax.bar("Gymnastics", mens_gymnastics["Height"].mean(), yerr= mens_gymnastics["Height"].std() )
ax.set_ylabel("Height (cm)")
plt.show()

print("----Metodo .errorbar() para graficar una serie de tiempo + marcas de error")
fig, ax = plt.subplots()
ax.errorbar(seattle_weather["MONTH"],
           seattle_weather["MLY-TAVG-NORMAL"],
           yerr=seattle_weather["MLY-TAVG-STDDEV"])
ax.errorbar(austin_weather["MONTH"],
            austin_weather["MLY-TAVG-NORMAL"],
            austin_weather["MLY-TAVG-STDDEV"]
           )
ax.set_ylabel("Temperature (Farenheit)")
plt.show()


# ADDING BOXPLOTS
print("----Metodo .boxplot() para crear una grafica de caja y bigotes")
fig, ax = plt.subplots()
ax.boxplot( [mens_rowing["Height"], mens_gymnastics["Height"]] )
ax.set_xticklabels(["Rowing","Gymnastics"])
ax.set_ylabel("Height (cm)")
plt.show()

```

# Quantitative comparisons: scatter plots

```python
# INTRODUCING SCATTERPLOT
print("----Metodo .scatter() para crear una grafica de dispersion")
fig, ax = plt.subplots()
ax.scatter(climate_change["co2"], climate_change["relative_temp"])
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperatures (Celsius)")
plt.show()

# CUSTOMIZING SCATTER PLOTS
print("----Metodo .scatter() para crear una grafica de dispersion con etiquetas")
ochentas = climate_change["1980-01-01":"1989-12-31"]
noventas = climate_change["1990-01-01":"1999-12-31"]
fig, ax = plt.subplots()
ax.scatter(ochentas.index, ochentas["relative_temp"], color="red", label= "ochentas")
ax.scatter(noventas.index, noventas["relative_temp"], color="blue", label="noventas")
ax.legend()
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperatures (Celsius)")
plt.show()

# ENCODING A THIRD VARIABLE AS COLOR
print("----Metodo .scatter() para agregar una tercera variable como color")
fig, ax = plt.subplots()
ax.scatter(climate_change["co2"], climate_change["relative_temp"], c=climate_change.index)
ax.set_xlabel("CO2 (ppm)")
ax.set_ylabel("Relative temperatures (Celsius)")
plt.show()
```


# Automating figures from data (Video)

```python
# GETTING UNIQUE VALUES OF A COLUMN OF A PANDAS DATA FRAME
print("----Metodo .unique() para obtener los valores unicos de una Serie de Pandas")
print( summer_2016_medals["Sport"].unique() )

# BAR CHAR OF HEIGHTS OF ALL SPORTS
print("----GRAFICA DE BARRAS DENTRO DE UN CICLO")
sports = summer_2016_medals["Sport"].unique()
fig, ax = plt.subplots()
for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
    ax.bar(sport,
           sport_df["Height"].mean(),
           yerr = sport_df["Height"].std()
          )
ax.set_xticklabels(summer_2016_medals["Sport"], rotation=90)
ax.set_ylabel("Height (cm)")
plt.show()

```

# Automate your visualization

```python
print("----Valores unicos de la columna Sport del dataframe")
sports_column = summer_2016_medals["Sport"]
sports = sports_column.unique()

print(sports)

print("---Ciclo sobre los valores unicos de Sports")
fig, ax = plt.subplots()
for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
    ax.bar(sport, sport_df["Weight"].mean(), yerr=sport_df["Weight"].std() )
ax.set_xticklabels(sports, rotation = 90)
ax.set_ylabel("Weight")
plt.show()
fig.savefig("sports_weights.png")
```
